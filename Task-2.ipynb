## TASK 2: Prediction using Decision Tree Classifier

 


Creating a Decision Tree Classifier on Iris Dataset and visualizing it graphically.

## Importing the Libraries

```python
# Importing all the required libraries.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing Dataset

```python
# Importing the data from CSV file.

df=pd.read_csv("Iris.csv")
df.head()
```

```python
# Setting the Index Column as "Id"

df=pd.read_csv("Iris.csv",index_col=0)
df.head()
```

```python
# Dimensions of the Dataframe

df.shape
```

```python
# Checking for any null values in the Dataframe

df.isnull().sum()
```

```python
# Checking the number of obsrvations for each distinct species present in the dataframe.

df["Species"].value_counts()
```

```python
# Creating a countplot of Species using Seaborn Library

import seaborn as sns
sns.countplot(x=df["Species"],palette="rainbow")
plt.title("Distribution for each distinct species")
plt.show()
```

## Data Preparation

```python
# Separating the Features and Target Variables into X and Y dataframes to further use it for training and testing data.
# Features

X=df.drop("Species",axis=1)
X.head()
```

```python
# Target Variable

Y=df[["Species"]]
Y.head()
```

```python
print("Features(Number of observations,Number of features) :",X.shape)
print("Target(Number of observations,Number of features) :",Y.shape)
```

## Splitting the Data into Training and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

## Fitting the Training Data into the Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(random_state=0).fit(X_train,Y_train)
```

## Visualizing the Decision Tree

```python
from sklearn import tree
features=["SepalLengthCm","SepalwidthCm","PetalLengthCm","PetalWidthCm"]
classes=["Iris-versicolor","Iris-setosa","Iris-virginica"]
fig=plt.figure(figsize=(15,10))
tree.plot_tree(dtree,feature_names=features,class_names=classes,filled=True);
```

## Evaluating the Accuracy of the Classifier

```python
from sklearn.metrics import accuracy_score
Y_predict=dtree.predict(X_test)
dtree_accuracy=accuracy_score(Y_test,Y_predict)
print("Accuracy of Classifier: {}%".format(round(dtree_accuracy*100,2)))
```
